```python
# 3. Inicjalizacja populacji
docKlient = aktywnosc.iloc[[2]] #docelowy klient - klient o ID: 3
docKlient

produkty_obejrzane = docKlient.loc[:, (docKlient != 0).any()]
produkty_obejrzane   

populacja_aktywnosc = aktywnosc.drop(columns = produkty_obejrzane)
populacja_kategorie = kategorie.drop(columns = produkty_obejrzane)

M = 150 # liczba osobnikow w populacji
n= 4 # liczba genow osobnika
populacja_osobnikow_kategorie = np.split(populacja_kategorie, M, axis=1)
populacja_osobnikow_aktywnosc = np.split(populacja_aktywnosc, M, axis=1)
```

```python
population = []
for x in range(len(populacja_osobnikow_kategorie)):
    population.append(populacja_osobnikow_kategorie[x].columns.values)                      
#population
population

#population
```

```python
# Konwersja binarna osobnikow
def vec_bin_array(arr, m):
    """
    Arguments: 
    arr: Numpy array of positive integers
    m: Number of bits of each integer to retain

    Returns a copy of arr with every element replaced with a bit vector.
    Bits encoded as int8's.
    """
    to_str_func = np.vectorize(lambda x: np.binary_repr(x).zfill(m))
    strs = to_str_func(arr)
    ret = np.zeros(list(arr.shape) + [m], dtype=np.int8)
    for bit_ix in range(0, m):
        fetch_bit_func = np.vectorize(lambda x: x[bit_ix] == '1')
        ret[...,bit_ix] = fetch_bit_func(strs).astype("int8")

    return ret 
popop = []
for x in range(M):
    popop.append(vec_bin_array(population[x],9))
#popop
```

```python
# Korelacja osobnikow

def Korelacja_osobnikow(populacja,populacja_binarnie):
    population_fitness_dictionary = dict([])
    for x in range(M):
        jac_sim = 1- pairwise_distances(populacja[x].T, metric = "hamming")
        # optionally convert it to a DataFrame
        jac_sim = pd.DataFrame(jac_sim, index=populacja[x].columns,
                           columns=populacja[x].columns)
        population_fitness_dictionary[str(populacja_binarnie[x])] = jac_sim.sum().sum()
    return population_fitness_dictionary
population_fitness_dictionary = Korelacja_osobnikow(populacja_osobnikow_kategorie,population)
population_fitness_dictionary
```

```python
# 5. Selekcja metoda koła ruletki

def get_probability_list(dictionary):
    fitness = dictionary.values()
    total_fit = float(sum(fitness))
    relative_fitness = [f/total_fit for f in fitness]
    probabilities = [sum(relative_fitness[:i+1]) 
                     for i in range(len(relative_fitness))]
    return probabilities
#prob = get_probability_list(population_fitness_dictionary)

def roulette_wheel_pop(population, probabilities, number):
    chosen = []
    for n in range(number):
        r = random.random()
        for (i, individual) in enumerate(population):
            if r <= probabilities[i]:
                chosen.append(list(individual))
                break
    return chosen
```

```python
# 6. Krzyzowanie 
def crossover4(a, b, crossover_point):
    a1 = a[:]
    b1 = b[:]
    for i in range(crossover_point):
        a1[i], b1[i] = b1[i], a1[i]
    return [a1, b1]
```

```python
#7. Mutacja
def mutation(pop_after_cross,mutation_rate):
    population_nextgen = []
    for i in range(0,len(pop_after_cross)):
        chromosome = pop_after_cross[i]
        for j in range(len(chromosome)):
            if random.random() < mutation_rate:
                chromosome[j]= not chromosome[j]
        population_nextgen.append(chromosome)
    #print(population_nextgen)
    return population_nextgen
```

```python
# 8. Podobienstwo uzytkownikow
def Podobienstwo1(populacja,populacja_binarnie):
    population_fitness_dictionary1 = dict([])
    for x in range(M):
        pearson_sim = 1- pairwise_distances(populacja[x].T, metric = "correlation")
        # optionally convert it to a DataFrame
        pearson_sim = pd.DataFrame(pearson_sim, index=populacja[x].columns,
                           columns=populacja[x].columns)
        population_fitness_dictionary1[str(populacja_binarnie[x])] = pearson_sim.sum().sum()
    return population_fitness_dictionary1
population_fitness_dictionary1 = Podobienstwo1(populacja_osobnikow_kategorie,population)
population_fitness_dictionary1
```

```python
liczba_generacji = 200 # liczba generacji
M = 150 # liczba osobnikow w populacji
n= 4 # liczba produktow w osobniku
p_mutacji = 0.1 # Prawdopodobienstwo mutacji
rand1 = random.randint(1,3)
rand2 = random.randint(1,8)
for i in range(liczba_generacji):
    wynik_korelacji = Korelacja_osobnikow(populacja_osobnikow_kategorie,popop)
    prob = get_probability_list(wynik_korelacji)
    wynik_r1 = roulette_wheel_pop(popop,prob,M)
    crossover4(wynik_r1[rand2][rand1],wynik_r1[rand2][rand1],3)
    mutation(wynik_r1[rand2],p_mutacji)   
    wynik_podobienstwa = Podobienstwo1(populacja_osobnikow_aktywnosc,wynik_r1)    
    prob = get_probability_list(wynik_podobienstwa)
    wynik_r2 = roulette_wheel_pop(wynik_r1,prob,M)        

posortowane_wyniki = {k: v for k, v in sorted(wynik_podobienstwa.items(), key=lambda item: item[1], reverse = True)}
posortowane_wyniki

```
